```python
import pdfplumber

pdf_path = "/content/drive/MyDrive/UPWORK/Document Analysis Using LLMs/eBook-How-to-Build-a-Career-in-AI.pdf"

output_path = "/content/drive/MyDrive/UPWORK/Document Analysis Using LLMs/output.txt"

with pdfplumber.open(pdf_path) as pdf:
  extracted_text = ""

  for page in pdf.pages:
    extracted_text += page.extract_text()

with open(output_path, "w") as text_file:
   text_file.write(extracted_text)

print(f"Text extracted and saved to {output_path}")
```

```python
with open(output_path, "r") as text_file:
  text = text_file.read()

print(text[:1000])
```

```python
# summarization
from transformers import pipeline




summarizer = pipeline("summarization", model="t5-small")
summary = summarizer(text, max_length=256, min_length=30, do_sample=False)
print(summary)
```

```python
summary[0]["summary_text"]
```

```python
# Split the Document into Sentences
import nltk
from nltk.tokenize import sent_tokenize

nltk.download("punkt")
nltk.download('punkt_tab')

sentences = sent_tokenize(text)

print(sentences[:10])
```

```python
len(sentences)
```

```python

# combine sentences into passages consisting of 200 words
passages = []
current_passage = ""

for sentence in sentences:
    if len(current_passage.split()) + len(sentence.split()) < 200:  # adjust the word limit as needed
        current_passage += " " + sentence
    else:
        passages.append(current_passage.strip())
        current_passage = sentence
if current_passage:
    passages.append(current_passage.strip())

print(passages[0])
```

```python
# Generate Questions from the Passages Using LLMs
qg_pipeline = pipeline("text2text-generation", "valhalla/t5-base-qg-hl")




```

```python
print(passages[25])
```

```python
qg_pipeline(passages[25])
```

```python
# function to generate questions using the pipeline
def generate_questions_pipelines(passage, min_question=3):
  input_text = f"PASSAGE: {passage}"
  results = qg_pipeline(input_text)
  questions = results[0]["generated_text"].split("<sep>")

  # ensure string is not empty
  questions = [q.strip() for q in questions  if q.strip()]

  if len(questions) < min_question:
    passage_sentences = passage.split(". ")

    for i in range(len(passage_sentences)):
      if len(questions) >= min_question:
        break

      additional_input = " ".join(passage_sentences[i:i+2])
      additional_results = qg_pipeline(f"generate questions: {additional_input}")
      additional_questions = additional_results[0]["generated_text"].split("<sep>")

      questions.extend([q.strip() for q in additional_questions if q.strip()])


  return questions[:min_question]

  # generate questions from passages

for idx,  passage in enumerate(passages):
  questions = generate_questions_pipelines(passage)
  print(f"Passage {idx+1}:\n{passage}\n")
  print("Generated Questions:")
  for question in questions:
    print(f"- {question}")
  print(f"\n{'*'*50}\n")
```

```python
# Answer the Generated Questions Using a QA Model

qa_pipeline = pipeline("question-answering", model="deepset/roberta-base-squad2")
```

```python
qa_pipeline("What is a language for starting your AI job search?", "Python is a language for starting your AI job search.")
```

```python
def answer_unique_questions(passages, qa_pipeline):
  answered_questions = set()

  for idx, passage in enumerate(passages):
    questions = generate_questions_pipelines(passage)

    for question in questions:
      if question not in answered_questions:
        answer = qa_pipeline({"question": question, "context": passage})
        print(f"Question: {question}")
        print(f"Answer: {answer['answer']}\n")
        answered_questions.add(question)

    print(f"{'*'*50}\n")
answer_unique_questions(passages, qa_pipeline)
```
