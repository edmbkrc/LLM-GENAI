# Llama Index: A library for answering complex questions based on data

Llama Index is a software library for indexing and efficiently querying large amounts of data, often in documents. Especially when dealing with large data sets, indexing techniques are vital for retrieving and analyzing data quickly. Llama Index is a toolkit developed with such needs in mind.

#### Basic Concepts and Components
**1. Data Indexing:**
   - Indexing data involves creating data structures to provide quick access to specific parts of a large data set. This enables faster querying and data operations.
   
**2. Reverse Indexing:**  
   - An inverse index is a data structure that shows which documents in a document collection contain certain terms. This is especially used in text search operations.

**Querying and Data Retrieval:**
   - Indexes provide quick access to data. This helps to greatly reduce the response friction of querying. Llama Index optimizes these processes and enables users to perform fast and accurate queries on large data sets.

#### Features and Uses

**1. Fast Search and Retrieval:**  
   - Llama Index is optimized to perform fast search and retrieval operations on large data sets. Even complex queries can be answered very quickly.

**2. Scalability:**  
   - Scalability is a critical factor when working with large data sets. Llama Index is designed to ensure that performance does not degrade as the data set size increases.

**3. Flexibility:**  
   - It can work with many different data types and formats. Whether it is text, numeric data or other data formats, Llama Index can index various data sources with its flexible structure.

**4. API and Integrations:**  
   - It provides APIs for use with various programming languages, making it easy to integrate with existing systems. This allows Llama Index to be easily adopted and used with different applications.

#### Technical Details

**1. Architecture:**  
   - Llama Index typically uses a client-server architecture. The client generates queries and sends them to the server, while the server processes the queries and returns the results. This architecture ensures high performance even in distributed systems.

**2. Index Types:**  
   - Various data structures such as inverted indexes, tree structures (e.g. B-trees, AVL trees), hash tables, etc. can be used.

**Document Scoring and Ranking:** 3.  
   - Various scoring algorithms can be used to determine the relevance of the requested documents. This is especially important in text search and information retrieval.

**4. Data Updates:**  
   - Indexes can be continuously updated; added, modified or deleted data can be dynamically reflected in the index.

**1. Search Engines:**  
   - A large amount of web pages can be indexed to provide faster and more accurate search results.

**2. E-commerce Sites:**  
   - By indexing product catalogs, users can quickly find the products they are looking for.

**3. Data Science and Analytics:**  
   - Indexing is used for fast and efficient analysis of large data sets.

**Enterprise Applications:**  
   - Fast searches can be performed on documents, e-mails or other corporate data.

## Download Video


```python
def download_video(url, output_path):
    """
    Download a video from a given url and save it to the output path.

    Parameters:
    url (str): The url of the video to download.
    output_path (str): The path to save the video to.

    Returns:
    dict: A dictionary containing the metadata of the video.
    """
    #from pytube import YouTube
    from pytubefix import YouTube
    from pytubefix.cli import on_progress

    yt = YouTube(url, on_progress_callback = on_progress) # youtube nesnesi oluşturularak indirilmek istenen url ile nesne ilişkilendirilir.
    metadata = {"Author": yt.author, "Title": yt.title, "Views": yt.views} # meta verileri çekilir.
    yt.streams.get_highest_resolution().download(                          #Videonun mevcut olan en yüksek çözünürlüğünü seçer.
                                                  output_path=output_path,
                                                  filename="input_vid.mp4" #output_path dizinine "input_vid.mp4" dosya adıyla indirir.
    )
    return metadata

    #yt.streams.get_by_resolution(resolution="360p")
```

## Video Processing



```python
def video_to_images(video_path, output_folder):
    import os
    from moviepy.editor import VideoFileClip
    """
    Convert a video to a sequence of images and save them to the output folder.

    Parameters:
    video_path (str): The path to the video file.
    output_folder (str): The path to the folder to save the images to.

    """
    clip = VideoFileClip(video_path)
    # VideoFileClip(video_path), verilen video dosyasını açar ve bir video klip nesnesi oluşturur. Bu nesne, videonun karelerine
    # ve seslerine erişmeyi ve onları işleyip kaydetmeyi sağlar.
    clip.write_images_sequence( #Bu yöntem, video klibinden alınan kareleri belirli bir dizine kaydeder.
        os.path.join(output_folder, "frame%04d.png"), fps=0.2 # configure this for controlling frame rate.
    )
```

## Audio Processing


```python
def video_to_audio(video_path, output_audio_path):
    from moviepy.editor import VideoFileClip
    """
    Convert a video to audio and save it to the output path.

    Parameters:
    video_path (str): The path to the video file.
    output_audio_path (str): The path to save the audio to.

    """
    clip = VideoFileClip(video_path)
    audio = clip.audio # video dosyasının ses parçasını audio değişkenine atar.
    audio.write_audiofile(output_audio_path) #lde edilen ses parçasını belirtilen output_audio_path yoluna kaydeder.
```

## Transformation audio to text


```python
def audio_to_text(audio_path):
    import speech_recognition as sr #Ses tanıma işlemleri için kullanılan popüler bir Python kütüphanesi.
    """
    Convert an audio file to text.

    Parameters:
    audio_path (str): The path to the audio file.

    Returns:
    test (str): The text recognized from the audio.

    """
    recognizer = sr.Recognizer() # Recognizer sınıfından bir nesne oluşturur. Bu nesne, ses tanıma işlemleri için kullanılır.
    audio = sr.AudioFile(audio_path) #AudioFile sınıfı, belirtilen ses dosyasını açar ve bir ses dosya nesnesi oluşturur.

    with audio as source:
        # Record the audio data
        audio_data = recognizer.record(source) #Ses dosyasından ses verilerini kaydeder ve audio_data nesnesine atar.

        try:
            # Recognize the speech
            text = recognizer.recognize_whisper(audio_data) #audio_data nesnesinden ses verilerini tanır ve metne dönüştürür.
                                                            #Bu işlem, Whisper modeli kullanarak yapılır.
        except sr.UnknownValueError: #Ses tanıma işlemi sesi anlayamazsa bu hatayı döndürür ve kullanıcıya sesin anlaşılamadığı bildirilir.
            print("Speech recognition could not understand the audio.")
        except sr.RequestError as e: #Tanıma hizmetine istek gönderilemezse bu hatayı döndürür ve hata mesajı yazdırılır.
            print(f"Could not request results from service; {e}")

    return text
```

## Building the Multi-Modal Index and Vector Store

This code creates a multimodal index using Llama Index, combining text and image data from a specified folder and storing the index in a LanceDB vector database.

Here's a step-by-step explanation:

1. **Import Necessary Modules:**
   - `MultiModalVectorStoreIndex` from `llama_index.core.indices`: Used for creating a multimodal index.
   - `SimpleDirectoryReader`, `StorageContext` from `llama_index.core`: Used for loading data and configuring storage.
   - `LanceDBVectorStore` from `llama_index.vector_stores.lancedb`: Used to interface with the LanceDB vector database.

2. **Initialize Vector Stores:**
   - `text_store = LanceDBVectorStore(uri="lancedb", table_name="text_collection")`: Creates a LanceDB vector store for text data, connecting to the "lancedb" database and using the table "text_collection".
   - `image_store = LanceDBVectorStore(uri="lancedb", table_name="image_collection")`: Creates another LanceDB vector store for image data, also connecting to "lancedb" but using the "image_collection" table.

3. **Configure Storage Context:**
   - `storage_context = StorageContext.from_defaults(vector_store=text_store, image_store=image_store)`: Creates a `StorageContext` object using the initialized text and image vector stores. This context defines where the index data will be stored.

4. **Load Documents:**
   - `documents = SimpleDirectoryReader(output_folder).load_data()`: Loads data (text and image files) from the `output_folder` directory using `SimpleDirectoryReader`.

5. **Create Multimodal Index:**
   - `index = MultiModalVectorStoreIndex.from_documents(documents, storage_context=storage_context)`: Creates a `MultiModalVectorStoreIndex` from the loaded `documents` and using the configured `storage_context`. This builds the actual index, linking text and image data based on their relationships, and stores the vector embeddings in LanceDB.

**In essence, this code sets up a system to create a searchable index of text and image data, enabling you to query the index using either text or images and retrieve related information from both modalities.** LanceDB is used as the vector database to store and manage the vector embeddings for efficient retrieval.


```python
from llama_index.core.indices import MultiModalVectorStoreIndex #Hem metin hem de görüntü verilerini içerebilen çok modelli bir vektör indeksi
                                                                # oluşturmak için kullanılır.
from llama_index.core import SimpleDirectoryReader, StorageContext
from llama_index.vector_stores.lancedb import LanceDBVectorStore
#SimpleDirectoryReader: Bir dizindeki dosyaları okuyup veri yüklemek için kullanılır.
#StorageContext: Vektör storeları yönetmek için kullanılan bir bağlam sağlar.
#LanceDBVectorStore: Vektör verilerini LanceDB veritabanında saklamak için kullanılır.



text_store = LanceDBVectorStore(uri="lancedb", table_name="text_collection") #Metin verilerini saklayacak bir LanceDB vektör deposu oluşturur
image_store = LanceDBVectorStore(uri="lancedb", table_name="image_collection") #Görüntü verilerini saklayacak bir LanceDB vektör deposu oluşturur.
storage_context = StorageContext.from_defaults(
    vector_store=text_store,
    image_store=image_store
)
# storage_context: text_store ve image_store vektör depolarını içeren bir saklama bağlamı oluşturur.
# Bu bağlam, vektör verilerinin nerede ve nasıl saklanacağını belirtir.

# Create the MultiModal index
documents = SimpleDirectoryReader(output_folder).load_data()
#output_folder dizinindeki tüm dosyaları okur. ve load_data() metodu ile bu dizindeki tüm verileri documents değişkenine yükler.

index = MultiModalVectorStoreIndex.from_documents(
    documents,                      # Bu değişkende bulunan verilerden (metin ve görüntü verileri) bir çok modelli vektör indeksi oluşturur.
    storage_context=storage_context #indeksin hangi vektör depolarını kullanacağını belirtir.
)
```

```python
from llama_index.core.response.notebook_utils import display_source_node
from llama_index.core.schema import ImageNode


def retrieve(retriever_engine, query_str):
    retrieval_results = retriever_engine.retrieve(query_str)

    retrieved_image = []
    retrieved_text = []
    for res_node in retrieval_results:
        if isinstance(res_node.node, ImageNode):
          #gözlemin imagenode formatında olup olmadığı kontrol edilir. Eğer imagenode formatındaysa true döner ve gözlemi  retrieved_image listesine atar.
            retrieved_image.append(res_node.node.metadata["file_path"])
        else:
            #display_source_node(res_node, source_length=200) # metin kaynağın ilk 200 karakteri ekranda görüntülenir.
            retrieved_text.append(res_node.text)

    return retrieved_image, retrieved_text
```

```python
# show the images

from PIL import Image
import matplotlib.pyplot as plt
import os


def plot_images(image_paths):
    images_shown = 0 # subplot konumları için kullanılacak
    plt.figure(figsize=(16, 9))
    for img_path in image_paths:
        if os.path.isfile(img_path): #image yolunun geçerli bir dosya olup olmadığı kontrol edilir.
            image = Image.open(img_path) # eğer geçerli bir dosya ise resim dosyası açılır

            plt.subplot(2, 3, images_shown + 1) # subplot konumları belirlenir
            plt.imshow(image)
            plt.xticks([])
            plt.yticks([])
            #`plt.xticks([])` ve `plt.yticks([])` ifadeleri ile grafik üzerinde eksen tikleri gizlenir.

            images_shown += 1
            if images_shown >= 7:
                break
```

```python
def video_RAG_chatbot(query_str, metadata_str):
    from llama_index.multi_modal_llms.openai import OpenAIMultiModal
    openai_mm_llm = OpenAIMultiModal(
        model="gpt-4o-mini",
        max_new_tokens=1500,
        temperature=0.0,
        top_p=1.0
    )

    img, txt=retrieve(retriever_engine=retriever_engine,query_str=query_str)
    image_documents = SimpleDirectoryReader(
        input_dir=output_folder,
        input_files=img).load_data()

    context_str = " ".join(txt)

    response = openai_mm_llm.complete(
        prompt=qa_tmpl_str.format(
            context_str=context_str,
            query_str=query_str,
            metadata_str=metadata_str
        ),
        image_documents=image_documents,
    )

    return response.text
```
